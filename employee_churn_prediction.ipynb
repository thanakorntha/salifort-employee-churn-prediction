### Methodology

This project leverages supervised machine learning, which is particularly suitable for this task as it can learn from labeled data (employee characteristics and their corresponding churn status) to predict future employee leaving. The project implements four popular binary classification algorithms: *Logistic Regression*, *Decision Tree*, *Random Forests*, and *XGBoost*. These models are chosen due to their effectiveness in classification tasks and their ability to handle various data types commonly found in HR datasets.

The project follows a five-step approach to achieve its goals:

1.  **Data Preparation and Cleaning:** This stage ensures data consistency and compatibility with the models. It involves handling missing values, outliers, and categorical data.
2.  **Data Analysis:** Exploratory Data Analysis (EDA) is performed to uncover relationships between employee characteristics and churn. This helps identify the most relevant features for model training.
3.  **Model Training and Fine-Tuning:** All four models are trained using the prepared data. Cross-validation techniques are employed to evaluate their performance, leading to the selection of the top two competitors. These chosen models are further fine-tuned by optimizing their hyperparameters, which significantly impact model behavior.
4.  **Model Evaluation:** The top two models with optimized hyperparameters are evaluated on a testing set using metrics like *recall*, *precision*, *F1-score*, and *AUC*. Recall prioritizes correctly identifying departing employees, while precision focuses on the accuracy of positive predictions. F1-score balances these two aspects. AUC measures the model's ability to distinguish between employees who stayed and left. Considering these metrics, with a focus on recall, the model with the best overall performance is chosen as the final churn prediction model. Additionally, a confusion matrix and ROC curve are utilized for a comprehensive performance evaluation.
5.  **Conclusion and Recommendations:** The final model is presented, including an analysis of feature importance. Recommendations to reduce employee churn and improve retention are formulated based on the model's findings.

It shows that there are 824 rows containing outliers in the ***`tenure`*** variable. The above boxplot shows the right-skewed distribution of the ***`tenure`*** variable. In other words, it is more likely that employees who have worked for more than 5.5 years in the company are outliers.

#### Training and Evaluating Candidate Models

In machine learning, defining model instances and evaluation metrics is crucial for training and evaluating candidate models. This process allows us to assess the performance of each model. The following details is a breakdown of the key components:

- **Model Instances** ***(`model_instances`)***: This variable stores instances of different machine learning models along with their descriptive names. This project utilizes four models: *logistic regression*, *decision trees*, *random forest*, and *XGBoost*.
- **Evaluation Metrics** ***(`scoring`)***: This variable holds the evaluation metrics to use to evaluate model performance. These metrics include *recall*, *precision*, *F1-score*, and *ROC-AUC*.
- **Evaluation Results** ***(`model_results`)***: This initially empty dictionary will store the evaluation results for each model after the training and evaluation process.

The construction and evaluation of each model occur within a loop. The ***`cross_validate()`*** function plays a vital role in this process. It generates cross-validated scores for each data point within the dataset. The function calculates all evaluation scores for each fold (a subset of the data used for training and validation) and then averages them to obtain the final evaluation score. Finally, these scores are stored in the ***`model_results`*** dictionary.

```python
# Create dictionary to store instances of different machine learning models along with their names
model_instances = {
    'LRC_BASE': LogisticRegression(random_state=random_num, max_iter=1000),
    'DTC_BASE': DecisionTreeClassifier(random_state=random_num),
    'RFC_BASE': RandomForestClassifier(random_state=random_num),
    'XGB_BASE': XGBClassifier(objective='binary:logistic', random_state=random_num)
}

# Define evaluation metrics to calculate
scoring = ['recall', 'precision', 'f1', 'roc_auc']

# Create empty dictionary to store the evaluation results of each model
model_results = {}

# Loop through each model instance
for model_name, model_instance in model_instances.items():
    
    # Perform cross-validation on current model
    scores = cross_validate(model_instance, X=X, y=y, cv=5, scoring=scoring)

    # Define dictionary that maps input metric to actual metric name in GridSearchCV
    metric_dict = {
        'recall': 'test_recall',
        'precision': 'test_precision',
        'f1': 'test_f1',
        'auc': 'test_roc_auc'
    }
    
    # Calculate mean evaluation scores of each metric across all folds
    mean_scores = {metric_name: np.mean(scores.get(metric_key)) for metric_name, metric_key in metric_dict.items()}
    
    # Store mean evaluation scores for current model
    model_results[model_name] = mean_scores
```

The model evaluation results are now displayed in a DataFrame.

```python
# Convert dictionary of evaluation results into DataFrame
model_results_df = pd.DataFrame(model_results).T.reset_index().rename(columns={'index': 'model'})

# Print model results
print(model_results_df.to_string(index=False))
```

#### Random Forest

Random Forest is an ensemble learning algorithm that builds upon the concept of bagging. It involves training multiple decision trees on bootstrapped data, meaning each tree is trained on a random sample with replacement from the original dataset. This introduces diversity among the trees, reducing overfitting. Beyond bootstrapping, Random Forest adds another layer of randomness by randomly selecting a subset of features at each split point within each tree. This further decorrelates the trees, preventing them from focusing on the same features and leading to a more robust model. While the individual trees in a random forest might be weak learners, their combined predictions through majority voting (for classification) or averaging (for regression) often lead to a more accurate and generalizable model. However, like decision trees, random forest models require hyperparameter tuning to optimize the number of trees, the maximum depth of each tree, and the number of features considered at each split for the best possible performance.

##### Hyperparameter Tuning

It is good to start dividing the dataset into training and testing sets, but outliers do not need to be be removed this time because random forest and XGBoost models are not sensitive to outliers. The ***`train_test_split()`*** function is used in order to split the dataset into a 70:30 ratio and stratify attribute to preserve the same proportions of observations in each set.

```python
# Split data into target (y) and feature (X) variables
y = hr_enc['left']
X = hr_enc.drop(columns=['left'], axis=1)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=random_num)

# Get shape of each training and testing set
print('Rows and columns of Training Set:', '`X` data', X_train.shape, 'and `y` data', y_train.shape)
print('Rows and columns of Testing Set:', '`X` data', X_test.shape, 'and `y` data', y_test.shape)
```

After splitting the dataset, the random forest model's hyperparameters are fine-tuned in order to achieve optimal performance. This involves training the model with various combinations of key settings and evaluating their effectiveness. Grid-search cross-validation is a common method for this task. It is important to explore different settings for following six hyperparameters:

-  ***`n_estimators`:*** Specifies the number of trees your model will build in its ensemble.
-  ***`max_depth`:*** Controls how deep each base learner tree will grow.
-  ***`min_samples_split`:*** Defines the minimum number of samples that a node must have to split into more nodes.
-  ***`min_samples_leaf`:*** Defines the minimum number of samples for a leaf node.
-  ***`max_features`:*** Specifies the number of features that each tree randomly selects during training.
-  ***`max_samples`:*** Defines the proportion of samples drawn from the training data to train each tree.

Recall score is used as the main evaluation metric, which focuses on the model's ability to correctly identify true positive cases. This approach ensures that the model is able to identify the optimal hyperparameter configuration for this task and then leads to the best possible model performance. Then, the model is fitted to the training data using the ***`fit()`*** method. The fitting time can vary depending on the complexity of the model and the size of the training set.

```python
# Instantiate random forest model
rfc = RandomForestClassifier(random_state=random_num)

# Define hyperparameters grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7, 9],
    'min_samples_split': [2, 3, 4, 5],
    'min_samples_leaf': [1, 2, 3],
    'max_features': [0.5, 0.7, 1.0],
    'max_samples': [0.5, 0.7, 1.0]
}

# Define list of scoring metrics
scoring = ['recall', 'precision', 'f1', 'roc_auc']

# Instantiate GridSearchCV object
rfc_cv = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5, scoring=scoring, n_jobs=-1, refit='recall')
```

```python
%%time

# Fit model to training set
rfc_cv = rfc_cv.fit(X_train, y_train)
```

After the model training is complete, the resulting model is saved as a pickle file, called ***`RFC_HYPE`***, in the ***`models`*** folder for later use.

```python
def write_pickle(path: str, model_object: object, model_name: str):
    """
    Save a model object using a pickle file.
    
    Params: path (str): The path where the pickle file will be saved
            model_object (object): The model object to be pickled and saved
            model_name (str): The filename for the pickle file (without the '.pkl' extension)
    
    Returns: None
    """
    
    # Define file path
    file_path = path + model_name + '.pkl'
    
    # Save pickle file
    with open(file_path, 'wb') as to_write:
        pickle.dump(model_object, to_write)
```

##### Model Validation

Then, it is time to evaluate how well the hyperparameters were optimized for the random forest model.

```python
def make_eval_results(models: dict, metric: str):    
    """
    Generate evaluation results for each model on the training set.
    
    Params: models (dict): A dictionary of model names and their corresponding objects.
            metric (str): The metric to evaluate the models on.
    
    Returns: DataFrame: A DataFrame containing the evaluation results for each model.
    """
    
    # Define list of evaluation results
    eval_results = []
    
    # Define dictionary that maps input metric to actual metric name in GridSearchCV
    metric_dict = {
        'recall': 'mean_test_recall',
        'precision': 'mean_test_precision',
        'f1': 'mean_test_f1',
        'auc': 'mean_test_roc_auc'
    }
    
    # Loop through each model
    for model_name, model_object in models.items():
        
        # Get all results from CV and put them in dataframe
        cv_results = pd.DataFrame(model_object.cv_results_)
        
        # Isolate row of `cv_results` with max(metric) score
        best_index = cv_results[metric_dict[metric]].idxmax()
        best_estimator_results = cv_results.iloc[best_index]
        
        # Extract metrics from row in `best_estimator_results`
        metrics = {metric_name: round(best_estimator_results.get(metric_key), 4) for metric_name, metric_key in metric_dict.items()}
        
        # Add `model` and `metrics` to `eval_results` list
        eval_results.append({'model': model_name, **metrics})
    
    # Return dataframe of evaluation results
    return pd.DataFrame(eval_results)
```

#### Predictive Model Performance

First, it is good to define a function to load a model object from a pickle file.

```python
def read_pickle(path: str, saved_model_name: str):
    """
    Load a model object from a pickle file.
    
    Params: path (str): The path where the pickle file is located
            saved_model_name (str): The filename of the pickle file (without the '.pkl' extension)
    
    Returns: Object: The Python object loaded from the pickle file.
    """
    
    # Define file path
    file_path = path + saved_model_name + '.pkl'
    
    # Load pickle file
    with open(file_path, 'rb') as to_read:
        model = pickle.load(to_read)
        
    # Return model
    return model
```

The previous function can be used to load the trained models from the ***`models`*** folder. Each model is saved to a dictionary named ***`final_models`*** by model name, as follow: ***`RFC_HYPE`*** for Random Forest model and ***`XGB_HYPE`*** for XGBoost model.

```python
# Create empty dictionary to store model names and fitted objects
final_models = {}

# Define pickle path
pickle_path = './models/'

# Load pickle fils
final_models['RFC_HYPE'] = read_pickle(pickle_path, 'RFC_HYPE')
final_models['XGB_HYPE'] = read_pickle(pickle_path, 'XGB_HYPE')
```

Then, it is important to define a function to get the prediction scores on the testing set for each model.

```python
def get_test_scores(models: dict, X_test_data, y_test_data):
    """
    Compute and return test scores for each model on the testing set.

    Params: models (dict): A dictionary of model names and their corresponding objects. 
            X_test_data: array-like of shape of X_test data.
            y_test_data: array-like of shape of y_test data.

    Returns: DataFrame: A DataFrame containing the test scores for each model.
    """

    # Define list of testing scores
    test_scores = []

    # Loop through each model
    for model_name, model_object in models.items():
        
        # Get predictions on testing set based on best parameters
        preds = model_object.best_estimator_.predict(X_test_data)
        
        # Compute and add testing scores to 'test_scores' list
        test_scores.append({
            'model': model_name,
            'recall': round(recall_score(y_test_data, preds), 4),
            'precision': round(precision_score(y_test_data, preds), 4),
            'f1': round(f1_score(y_test_data, preds), 4),
            'auc': round(roc_auc_score(y_test_data, preds), 4)
        })

    # Return dataframe of testing scores
    return pd.DataFrame(test_scores)
```

The ***`get_test_scores()`*** function that is defined in earlier step can be used to get the prediction scores on the testing set for each model and then store them in a dataframe, called ***`test_scores_df`***. This dataframe can be used to compare the performance of the models on the testing set.

```python
# Create dataframe to store 'test_scores' list and assign to 'test_scores_df'
test_scores_df = get_test_scores(final_models, X_test, y_test)

# Display results for Random Forest and XGBoost models
print(test_scores_df)
```

Surprisingly, the result shows that recall score is the same for both models. This suggests that the Random Forest model is better at predicting whether an employee will leave the company.

#### ROC Curve and AUC

The Receiver Operating Characteristic (ROC) curve is a visual representation of a classification model's performance across all possible classification thresholds. It depicts the trade-off between two key metrics: True Positive Rate (TPR), also known as recall, which indicates the proportion of true positives correctly classified, and False Positive Rate (FPR), which represents the proportion of true negatives incorrectly classified as positives. It is another common tool used to evaluate the performance of a classification model. Ideally, a good ROC curve should hug the upper left corner, indicating high TPR and low FPR, meaning the model accurately identifies true positives while minimizing false positives.

```python
# Define ROC values for random forest model
rf_pred = final_models['RFC_HYPE'].predict(X_test)
rf_proba = final_models['RFC_HYPE'].predict_proba(X_test)[::,1]
rf_auc = roc_auc_score(y_test, rf_pred)
rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test, rf_proba)

# Define ROC values for XGBoost model
xgb_pred = final_models['XGB_HYPE'].predict(X_test)
xgb_proba = final_models['XGB_HYPE'].predict_proba(X_test)[::,1]
xgb_auc = roc_auc_score(y_test, xgb_pred)
xgb_fpr, xgb_tpr, xgb_thresholds = roc_curve(y_test, xgb_proba)

# Set figure size
plt.figure(figsize=(10, 5))

# Plot ROC curve for each model
plt.plot(rf_fpr, rf_tpr, label='RFC (AUC = %0.3f)' % rf_auc)
plt.plot(xgb_fpr, xgb_tpr, label='XGB (AUC = %0.3f)' % xgb_auc)
plt.plot([0, 1], [0, 1], 'r--') # Diagonal line

# Add axis labels and title
plt.title('Receiver Operating Characteristic (ROC)')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc='lower right')

# Display plot
plt.show()
```

Both the Random Forest and XGBoost models demonstrated strong performance based on their ROC curves. They effectively distinguished between true positives and false positives. This indicates that both models were able to correctly identify true positives while minimizing the number of false positives. However, the Random Forest model exhibited strong overall performance as measured by the Area Under the Curve (AUC). The AUC curve provides a single metric summarizing the overall performance. A higher AUC indicates a better model. In this case, the Random Forest model had an AUC of 94.8%, which is higher than the XGBoost model at 94.7%.

#### Review of Evaluation Metrics

A comprehensive model evaluation requires considering various metrics beyond a single score. This project utilizes the various metrics to evaluate amd compare the performance of the employee churn prediction models.

For ***confusion matrix***, there are four possible outcomes that can be observed:

-  **True Positive (TP):** Indicates a correct classification of a positive observation as positive class
-  **True Negative (TN):** Indicates a correct classification of a negative observation as a negative class
-  **False Positive (FP):** Indicates an observation belonging to the negative class being falsely classified as positive
-  **False Negative (FN):** Indicates an observation belonging to the positive class being falsely classified as negative

For ***predictive scores***, there are four main metrics that can be observed:

-  **Recall:** Measures the model's ability to identify employees who will leave (TP) out of all actual leavers. High recall is crucial when missing true positives (FN) is costly.
-  **Precision:** Measures the accuracy of positive predictions, indicating how many identified leavers (TP) were actual leavers. This is important when misclassifying employees who will stay (FP) can be detrimental.
-  **F1-Score:** Combines precision and recall into a single metric, providing a balanced view of the model's performance.
-  **AUC-ROC:** Represents the probability that the model will rank a random positive instance higher than a random negative instance. This metric is particularly useful when dealing with imbalanced datasets.

By analyzing these metrics together, we can gain a deeper understanding of the strengths and weaknesses of each model, allowing us to select the most suitable one for predicting employee churn at Salifort Motors.
